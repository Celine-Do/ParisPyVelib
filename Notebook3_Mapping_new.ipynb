## 2. Mapping

```python
#Il faut changer coord en mercator
import math

# derived from the Java version explained here: http://wiki.openstreetmap.org/wiki/Mercator
RADIUS = 6378137.0 # in meters on the equator

def lat2y(a):
    return math.log(math.tan(math.pi / 4 + math.radians(a) / 2)) * RADIUS

def lon2x(a):
    return math.radians(a) * RADIUS
```

```python
df_lat_lon.info()
#Changement de type de Latitude et Longitude
```

```python
df_lat_lon['Latitude'] = df_lat_lon['Latitude'].astype('float64')
df_lat_lon['Longitude'] = df_lat_lon['Longitude'].astype('float64')
```

```python
#Vérification
df_lat_lon.info()
```

```python
df_lat_lon['Latitude M'] = df_lat_lon['Latitude'].apply(lat2y)
df_lat_lon['Longitude M'] = df_lat_lon['Longitude'].apply(lon2x)
```

```python
df_lat_lon.head()
```

Pour une meilleure lecture avec Bokeh, il faut supprimer les apostrophes " ' "

```python
df_lat_lon['Address'] = df_lat_lon['Address'].str.replace(r"[\"\',]", '')
```

Définition des bornes

```python
print('Latitude min : ', df_lat_lon['Latitude M'].min())
print('Latitude max : ', df_lat_lon['Latitude M'].max())
print('Longitude min : ', df_lat_lon['Longitude M'].min())
print('Longitude max : ', df_lat_lon['Longitude M'].max())
```

```python
#On groupe le dataFrame pour avoir le nombre de comptage total par adresse
df2 = df_lat_lon.groupby(['Address','Latitude', 'Longitude', 'Latitude M', 'Longitude M']).agg({'Count_by_hour' : lambda Count_by_hour : Count_by_hour.count()}).reset_index()
df2 = df2.sort_values(by = ['Count_by_hour'],ascending=False, axis=0)
df2.head(60)
```

```python
df2.nunique()
```

```python
#Top 10 des addresses avec le plus passage total
#2018
df_2018 = df_lat_lon[(df_lat_lon['Source'] == 2018)]
df2018 = df_2018.groupby(['Address','Latitude', 'Longitude', 'Latitude M', 'Longitude M', 'Source']).agg({'Count_by_hour' : lambda Count_by_hour : Count_by_hour.count()}).reset_index()
df2018 = df2018.sort_values(by = ['Count_by_hour'],ascending=False, axis=0)
df2018 = df2018.head(10)


#2019
df_2019 = df_lat_lon[df_lat_lon.Source == 2019]
df2019 = df_2019.groupby(['Address','Latitude', 'Longitude', 'Latitude M', 'Longitude M', 'Source']).agg({'Count_by_hour' : lambda Count_by_hour : Count_by_hour.count()}).reset_index()
df2019 = df2019.sort_values(by = ['Count_by_hour'],ascending=False, axis=0)
df2019 = df2019.head(10)

#2020
df_2020 = df_lat_lon[df_lat_lon.Source == 2020]
df2020 = df_2020.groupby(['Address','Latitude', 'Longitude', 'Latitude M', 'Longitude M', 'Source']).agg({'Count_by_hour' : lambda Count_by_hour : Count_by_hour.count()}).reset_index()
df2020 = df2020.sort_values(by = ['Count_by_hour'],ascending=False, axis=0)
df2020 = df2020.head(10)

#2021
df_2021 = df_lat_lon[df_lat_lon.Source == 2021]
df2021 = df_2021.groupby(['Address','Latitude', 'Longitude', 'Latitude M', 'Longitude M', 'Source']).agg({'Count_by_hour' : lambda Count_by_hour : Count_by_hour.count()}).reset_index()
df2021 = df2021.sort_values(by = ['Count_by_hour'],ascending=False, axis=0)
df2021 = df2021.head(10)

df2018['Color']= 'indigo'
df2019['Color']= 'purple'
df2020['Color']= 'blueviolet'
df2021['Color']= 'magenta'
```

```python
from bokeh.tile_providers import get_provider
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.models.tools import HoverTool
from bokeh.models.widgets import Panel, Tabs

source = ColumnDataSource(df2)
source2018 = ColumnDataSource(df2018)
source2019 = ColumnDataSource(df2019)
source2020 = ColumnDataSource(df2020)
source2021 = ColumnDataSource(df2021)

hover = HoverTool(tooltips=[("Address", "@Address"), ("Lat", "@Latitude"), ("Lng", "@Longitude")])

p = figure(x_range = (250000, 270000), y_range = (6244414, 6257395), 
           x_axis_type = 'mercator', y_axis_type = 'mercator')
p2018 = figure(x_range = (250000, 270000), y_range = (6244414, 6257395), 
           x_axis_type = 'mercator', y_axis_type = 'mercator')
p2019 = figure(x_range = (250000, 270000), y_range = (6244414, 6257395), 
           x_axis_type = 'mercator', y_axis_type = 'mercator')
p2020 = figure(x_range = (250000, 270000), y_range = (6244414, 6257395), 
           x_axis_type = 'mercator', y_axis_type = 'mercator')
p2021 = figure(x_range = (250000, 270000), y_range = (6244414, 6257395), 
           x_axis_type = 'mercator', y_axis_type = 'mercator')

tuile = get_provider('OSM')
p.add_tile(tuile)
p2018.add_tile(tuile)
p2019.add_tile(tuile)
p2020.add_tile(tuile)
p2021.add_tile(tuile)

p.circle(x='Longitude M', y='Latitude M',source = source, fill_alpha =0.6, fill_color = 'blue',line_color=None, size = 15)
p2018.circle(x='Longitude M', y='Latitude M',source = source2018, fill_alpha =0.6, fill_color = 'Color',line_color=None, size = 15)
p2019.circle(x='Longitude M', y='Latitude M',source = source2019, fill_alpha =0.6, fill_color = 'Color',line_color=None, size = 15)
p2020.circle(x='Longitude M', y='Latitude M',source = source2020, fill_alpha =0.6, fill_color = 'Color',line_color=None, size = 15)
p2021.circle(x='Longitude M', y='Latitude M',source = source2021, fill_alpha =0.6, fill_color = 'Color',line_color=None, size = 15)

p.add_tools(hover)
p2018.add_tools(hover)
p2019.add_tools(hover)
p2020.add_tools(hover)
p2021.add_tools(hover)

tab = Panel(child = p, title = 'Carte de Paris')
tab2018 = Panel(child = p2018, title = 'top de 2018')
tab2019 = Panel(child = p2019, title = 'top de 2019')
tab2020 = Panel(child = p2020, title = 'top de 2020')
tab2021 = Panel(child = p2021, title = 'top de 2021')

tabs= Tabs(tabs=[tab, tab2018, tab2019, tab2020, tab2021])
show(tabs)
```
